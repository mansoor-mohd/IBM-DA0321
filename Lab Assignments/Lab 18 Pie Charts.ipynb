### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the dataset
df = pd.read_csv("survey-data.csv")

# Step 3: Process the DatabaseWantToWorkWith column
# Split the semicolon-separated values and create a list of all databases
databases = df['DatabaseWantToWorkWith'].str.split(';', expand=True).stack().str.strip()

# Step 4: Count the frequency of each database
database_counts = databases.value_counts()

# Step 5: Select the top 5 databases
top_5_databases = database_counts.head(5)

# Step 6: Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_databases, labels=top_5_databases.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the dataset
df = pd.read_csv("survey-data.csv")

# Step 3: Process the OpSysProfessional use column
# Count the frequency of each operating system, excluding missing values
os_counts = df['OpSysProfessional use'].dropna().value_counts()

# Step 4: Select the top 5 operating systems (or all if fewer than 5)
top_os = os_counts.head(5)

# Step 5: Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_os, labels=top_os.index, autopct='%1.1f%%', startangle=140)
plt.title('Top Operating Systems Used Professionally')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.show()
```
