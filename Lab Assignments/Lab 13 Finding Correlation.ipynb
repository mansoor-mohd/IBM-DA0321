##### Step 1: Install and Import Required Libraries


```python
# Install the necessary libraries
!pip install pandas
!pip install matplotlib
!pip install seaborn

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python

# Calculate the 1st and 99th percentiles to exclude extreme outliers
lower_limit = df['ConvertedCompYearly'].quantile(0.01)
upper_limit = df['ConvertedCompYearly'].quantile(0.99)


# Create the histogram, focusing on the main range
plt.figure(figsize=(10, 6))  # Set figure size for clarity
sns.histplot(df['ConvertedCompYearly'], kde=True, bins='auto', binrange=(lower_limit, upper_limit))
plt.title('Distribution of Yearly Compensation (excluding top and bottom 1%)')
plt.xlabel('Yearly Compensation')
plt.ylabel('Frequency')



plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
## Write your code here
# Filter the dataframe to include only full-time employees
full_time_df = df[df['Employment'] == 'Employed, full-time']

# Calculate the median of 'ConvertedCompYearly' for full-time employees
median_comp_full_time = full_time_df['ConvertedCompYearly'].median()

# Display the result
print(f"The median yearly compensation for full-time employees is: {median_comp_full_time}")
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
## Write your code here

# Count respondents per country
country_counts = df['Country'].value_counts()

# Select the top 10 countries
top_countries = country_counts.head(10).index

# Filter the dataset for these countries
df_top_countries = df[df['Country'].isin(top_countries)]

# Create the box plot
plt.figure(figsize=(12, 8))  # Larger size for clarity
sns.boxplot(x='Country', y='ConvertedCompYearly', data=df_top_countries)
plt.title('Yearly Compensation Distribution by Country (Top 10 Countries)')
plt.xlabel('Country')
plt.ylabel('Yearly Compensation')
plt.xticks(rotation=45)  # Rotate country names for readability
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
## Write your code here
# Calculate Q1 and Q3
Q1 = df['ConvertedCompYearly'].quantile(0.25)
Q3 = df['ConvertedCompYearly'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create a new DataFrame without outliers
df_no_outliers = df[(df['ConvertedCompYearly'] >= lower_bound) & (df['ConvertedCompYearly'] <= upper_bound)]

# Display the shape of the original and new DataFrame
print(f"Original dataset shape: {df.shape}")
print(f"New dataset shape without outliers: {df_no_outliers.shape}")
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
## Write your code here
# Import necessary libraries

import seaborn as sns
import matplotlib.pyplot as plt

# Select the relevant columns from the dataset
columns_of_interest = ['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']
df_selected = df_no_outliers[columns_of_interest]

# Calculate the correlation matrix
correlation_matrix = df_selected.corr()

# Create a heatmap to visualize the correlations
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap: ConvertedCompYearly, WorkExp, and JobSatPoints_1')
plt.show()
```

<h3>Step 8: Scatter Plot for Correlations</h3>


**Task**: Create scatter plots to examine specific correlations between `ConvertedCompYearly` and `WorkExp`, as well as between `ConvertedCompYearly` and `JobSatPoints_1`.


```python
## Write your code here

# Scatter plot for ConvertedCompYearly vs. WorkExp
plt.figure(figsize=(8, 6))
sns.scatterplot(x='WorkExp', y='ConvertedCompYearly', data=df_no_outliers)
plt.title('Scatter Plot: ConvertedCompYearly vs. WorkExp')
plt.xlabel('Work Experience (Years)')
plt.ylabel('Yearly Compensation')
plt.show()

# Scatter plot for ConvertedCompYearly vs. JobSatPoints_1
plt.figure(figsize=(8, 6))
sns.scatterplot(x='JobSatPoints_1', y='ConvertedCompYearly', data=df_no_outliers)
plt.title('Scatter Plot: ConvertedCompYearly vs. JobSatPoints_1')
plt.xlabel('Job Satisfaction Points')
plt.ylabel('Yearly Compensation')
plt.show()
```
